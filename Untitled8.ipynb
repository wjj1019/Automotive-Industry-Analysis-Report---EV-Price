```python
from PIL import Image
import pytesseract

# If you don't have tesseract executable in your PATH, include the following:
# pytesseract.pytesseract.tesseract_cmd = r'<full_path_to_your_tesseract_executable>'

def extract_text_from_image(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)
    return text

```

```python
image_path = 'testimage.png'
text = extract_text_from_image(image_path)
```

```python
import re

def parse_t4_text(text):
    # Pattern for SIN
    sin_pattern = r'Numéro d\'assurance sociale CPP/QPP El PPIP\s+—_ Employment code Employee\'s QPP contributions — line 308 CPP/QPP pensionable earnings\s+([\d ]+)'
    
    # Pattern for boxes (with an optional bracket-enclosed number)
    box_pattern = r'(\d+)\s+([\d,)\.]+\d)'
    
    sin_match = re.search(sin_pattern, text)
    sin_num = sin_match.group(1).replace(' ', '') if sin_match else None

    box_matches = re.findall(box_pattern, text)
    boxes_dict = {box_num: float(box_val.replace(',', '').replace(')','')) for box_num, box_val in box_matches}

    return boxes_dict, sin_num

boxes, sin = parse_t4_text(text)
print(f"SIN: {sin}")
print("Boxes:")
for box, value in boxes.items():
    print(f"Box {box}: {value}")
```

```python
import re

def parse_t4_text_extended(text):
    # Pattern for SIN
    sin_pattern = r'Numéro dassurance sociale ([\d\s]+)'
    sin_match = re.search(sin_pattern, text)
    sin_num = sin_match.group(1).replace(' ', '') if sin_match else None

    # Pattern for boxes (with an optional bracket-enclosed number)
    box_pattern = r'(\d+)\s+([\d,\.]+\d)'
    box_matches = re.findall(box_pattern, text)
    boxes_dict = {box_num: float(box_val.replace(',', '')) for box_num, box_val in box_matches}

    # Pattern for employer's name
    employer_pattern = r"Employers name - Nom de employer ([\w\s\d]+) Year"
    employer_match = re.search(employer_pattern, text)
    employer_name = employer_match.group(1) if employer_match else None

    # Pattern for employee's name
    employee_pattern = r"Lastname \(capt ltes\)—Nomde fale \(entetes motes\) __Fratname Prom _ttal-hidale \| \[\d+\] [\d,\.]+\d+ \d+ ([\w\s\d]+) RPP contributions"
    employee_match = re.search(employee_pattern, text)
    employee_name = employee_match.group(1) if employee_match else None

    # Pattern for address
    address_pattern = r"([0-9]{1,5} .+,.+ [A-Z0-9]{3} [A-Z0-9]{3}) Facteur"
    address_match = re.search(address_pattern, text)
    address = address_match.group(1) if address_match else None

    return {
        "sin": sin_num,
        "boxes": boxes_dict,
        "employer_name": employer_name,
        "employee_name": employee_name,
        "address": address
    }


parsed_data = parse_t4_text_extended(text)
print(f"SIN: {parsed_data['sin']}")
print("Boxes:")
for box, value in parsed_data['boxes'].items():
    print(f"Box {box}: {value}")
print(f"Employer's name: {parsed_data['employer_name']}")
print(f"Employee's name: {parsed_data['employee_name']}")
print(f"Address: {parsed_data['address']}")
```

```python
import cv2
import pytesseract

def extract_info(image_path, coords):
    # Load the image
    img = cv2.imread(image_path, 0)  # 0 flag for grayscale

    extracted_info = {}

    for field in coords:
        # Get the region of interest
        roi = img[coords[field][1]:coords[field][1] + coords[field][3], coords[field][0]:coords[field][0] + coords[field][2]]

        # Use pytesseract to do OCR on the cropped rectangle
        text = pytesseract.image_to_string(roi)

        # Clean up the text and store it in the dictionary
        extracted_info[field] = text.replace('\n', ' ').strip()

    return extracted_info


extracted_info = extract_info('testimage.png', coords)
for key, value in extracted_info.items():
    print(f"{key}: {value}")
```

```python
def template_matching(image_path, template_path):
    # Load the image and the template
    img = cv2.imread(image_path,0)
    template = cv2.imread(template_path,0)
    
    # If template is larger than image, resize template to match the image size
    if template.shape[0] > img.shape[0] or template.shape[1] > img.shape[1]:
        template = cv2.resize(template, (img.shape[1], img.shape[0]), interpolation = cv2.INTER_AREA)

    w, h = template.shape[::-1]

    # Perform template matching
    res = cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

    # Draw a rectangle around the matched region
    top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv2.rectangle(img,top_left, bottom_right, 255, 2)

    return img

matched_image = template_matching('testimage2.png', 'template.jpg')

# You can visualize it using matplotlib
import matplotlib.pyplot as plt
plt.imshow(matched_image, cmap='gray')
plt.show()
```
